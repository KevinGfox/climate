```python
# import libraries
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.offline import init_notebook_mode, iplot
```

```python
# import data
df = pd.read_csv("GlobalTemperatures.csv")
```

# EDA

```python
df.info()
print("")
df.describe(include='all')
```

```python
# Apply filter on date column
df = df[df.dt > "1850-01-01"]
```

```python
# Create new column from date column
df['year'] = pd.DatetimeIndex(df['dt']).year
```

```python
# create mean temp from prvious column
mean_temp = df.groupby('year')['LandAndOceanAverageTemperature'].mean()
```

```python
# Display graphic about average sea & land temperature with mean line
fig = go.Figure()
fig.add_trace(go.Scatter(x = df.dt, y = df.LandAndOceanAverageTemperature))
fig.add_trace(go.Scatter(x=mean_temp.index,y=mean_temp.values, mode='lines'))
fig.add_vrect(x0="2015-12-31", x1="1970-01-01", line_width=0, fillcolor="red", opacity=0.2)
fig.show()
```
